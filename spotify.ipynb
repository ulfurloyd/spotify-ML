# Clustering and Predicting Spotify Song Popularity Using Machine Learning

## Problem Statement
Music streaming platforms like Spotify and Apple Music play an integral role in the modern music industry, bringing new eyes to artists' music with numbers up to the billions.

This makes it a valuable effort to understand patterns in music and predict a song's popularity using machine learning practices.

## Objectives
- Analyze and understand song structure using audio features.
- Group similar songs using unsupervised learning (K-means clustering).
- Predict a song's popularity score using regression techniques.
- Visualize trends and patterns in the dataset to gain deeper insights.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Dataset Overview
The dataset is sourced from [Kaggle- Ultimate Spotify Tracks DB](https://www.kaggle.com/datasets/zaheenhamidani/ultimate-spotify-tracks-db).

It contains over 2,32,000 songs with various audio features such as:
- 'danceability', 'energy', 'valence', 'loudness', 'tempo', 'instrumentalness' etc.
- The target variable is 'popularity', which is a score ranging from 0 to 100 assigned by Spotify.

These features are extracted by Spotify's API and describe the general vibe of each song.

```python
df = pd.read_csv("spotify.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Data Preprocessing

Performing the following preprocessing steps:

- Remove irrelevant columns such as song names and IDs.
- Dropped duplicate and missing entries to ensure data quality.
- Scaled numeric features using 'StandardScaler' to bring them to a similar scale for better model performance.

```python
# Drop unnecessary columns
df = df.drop(columns=['track_id', 'track_name', 'artist_name', 'album_name', 'duration_ms'], errors='ignore')

# Remove duplicates and missing values
df = df.drop_duplicates()
df = df.dropna()

df.info()
```

## Exploratory Data Analysis (EDA)

Using visualization to identify interesting trends in the data:

- A heatmap to explore correlation between audio features.
- A scatter plot of 'energy' vs. 'valence' to understand how happy or energetic songs are.
- Distribution of features to spot outliers or skewed values.

    These insights help in selecting features and understanding music patterns.

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Model Selection and Implementation

Implementing two types of models:

- **K-means Clustering**: To group similar songs based on audio features.
- **Linear Regression**: To predict a song's popularity score using audio features as inputs.

The models were trained using scikit-learn, and standard practices like train_test_split and feature scaling were followed.

## Feature Scaling

Scaling the numerical features to ensure that all values are on a similar scale. This is especially important for clustering and regression models that are sensitive to feature meagnitude.

```python
from sklearn.preprocessing import StandardScaler

features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 
            'instrumentalness', 'liveness', 'valence', 'tempo']

X = df[features]

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Clustering Songs with K-means

Using K-means clustering to group songs based on their audio features. This helps to understand natural groupings and similarities in the dataset.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
df['Cluster'].value_counts()
```

## Visualizing Clusters

We plot the clusters based on energy and valence — two intuitive features related to how "exciting" or "happy" a song feels. Different colors represent different clusters.

```python
plt.figure(figsize=(20, 12))
sns.scatterplot(data=df, x='energy', y='valence', hue='Cluster', palette='Set2')
plt.title("KMeans Clustering of Songs (Energy vs Valence)")
plt.show()
```

## Predicting Song Popularity (Regression)

We train a linear regression model using audio features to predict the popularity of a track. This helps us understand what makes a song more likely to be popular.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Prepare train/test split
target = 'popularity'
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Train the model
reg = LinearRegression()
reg.fit(X_train, y_train)

# Predict
y_pred = reg.predict(X_test)
```

## Evaluation Metrics

- **Mean Squared Error (MSE)**: Measures average squared difference between predicted and actual popularity.
- **R2 Score**: Measures how well the features explain the variability of popularity.

These metrics help in quantifying the model's accuracy.

```python
from sklearn.metrics import mean_squared_error, r2_score

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

## Actual vs Predicted Popularity

This scatter plot shows how well our regression model performed by comparing the actual popularity scores with the predicted ones.

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='purple')
plt.xlabel("Actual Popularity")
plt.ylabel("Predicted Popularity")
plt.title("Actual vs Predicted Popularity")
plt.grid(True)
plt.show()
```

## Results and Interpretation

- The K-means clustering grouped songs into 4 distinct clusters, showing different musical profiles.
- The regression model showed a reasonable ability to predict popularity based on audio features.
- A scatter plot of actual vs. predicted values shows that while the predictions aren't perfect, there is a visible correlation.

This indicates that audio features can be partially predictive of a song's success.
