# Cancer Diagnosis Prediction Using Machine Learning

## Problem Statement

- Cancer is one of the most common and life-threatening conditions globally.
- Early diagnosis can significantly improve survival rates.
- However, accurate diagnosis requires expensive and invasive procedures.

The goal of this project is to develop a ML model that can help predict whether a tumor is cancerous or not based on various medical features.

## Project Objectives

- To clean and explore the cancer dataset.
- To visualize data and understand patterns.
- To predict the diagnosis (malignant or benign) using ML models.
- To evaluate model performance using accuracy and classification metrics.

```python
import pandas as pd 
import seaborn as sns 
```

## Dataset Overview

The dataset is obtained from Kaggle.

It includes 30+ numerical features that describe characteristics of the cells present in cancer biopsies.

Key features include:
- radius_mean, texture_mean, area_mean, etc.

The target variable is `diagnosis` (M = Malignant, B = Benign).

## Data Preprocessing

- Loaded the dataset using `pandas`.
- Identified and removed unnecessary columns like `Unnamed: 32`.
- Converted the `diagnosis` column to binary data (1 for Malignant, 0 for Benign).
- Checked for null values and duplicates.
- Performed standardization before training models.

```python
data = pd.read_csv("cancerDataset.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
 data.describe()
```

```python
  sns.heatmap(data.isnull()) 
```

```python
data.drop(["Unnamed: 32","id"],axis = 1, inplace=True)
data.head()
```

```python
data.diagnosis =[ 1 if value == "M" else 0 for value in data.diagnosis]

    
```

```python
data["diagnosis"]
```

```python
data["diagnosis"] = data["diagnosis"].astype("category", copy=False)
data["diagnosis"].value_counts().plot(kind="bar")


# data.info()
```

```python
# divide into target variable and predictos 
y = data["diagnosis"] # our target variable
x = data.drop(["diagnosis"], axis=1) #our predicted variable
```

```python
y
```

```python
x
```

## Model Selection and Implementation

Implemented two models:

- **Logistic Regression**: For binary classification of cancer diagnosis.
- **K-means Clustering**: To group data without labels and compare clusters to real diagnoses.

```python
 from sklearn.preprocessing import StandardScaler
# create a scaler object
scaler = StandardScaler()

# fit he scaler to the data and transfrom the data
X_scaled = scaler.fit_transform(x)
x
```

```python
 X_scaled
# each row has 0 mean , each row has 1 standard deviation
```

### Split the data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test =  train_test_split(X_scaled,y,test_size=0.30, random_state=42) 
```

### Train the model

```python
from sklearn.linear_model import LogisticRegression

# create the lr model
lr = LogisticRegression()

# train the model on the training data
lr.fit(x_train, y_train)

# predict the target variable on the test data
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

## Evaluation Metrics

- **Logistic Regression** was evaluated using:
    - Accuracy Score
    - Confusion Matrix
    - Classification Report (Precision, Recall, F1-score)
- **K-means Clustering** was evaluated using label comparison and accuracy estimation.

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report
rep = classification_report(y_test, y_pred)
```

```python
print(rep)
```

## Results and Interpertation

- Logistic Regression showed high accuracy and well-balanced precision/recall
- K-means clustering provided decent separation between the classes but wasn't as accurate as supervised learning.
